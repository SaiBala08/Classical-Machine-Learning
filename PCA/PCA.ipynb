```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
print(X.shape)
plt.scatter(X[:, 0], X[:, 1])
plt.axis('equal');
```

```python
C = np.cov(X)
U, S, V = np.linalg.svd(C, full_matrices=True)
K = 1
print(V.shape)
```

```python
U_reduce = U[:,0:K];
Z = X.T@U_reduce;
print(Z.shape)
print(U_reduce.shape)
```

```python
X_rec = Z * U_reduce.T

plt.scatter(X_rec[0, :], X_rec[1, :])
plt.axis('equal');
```
